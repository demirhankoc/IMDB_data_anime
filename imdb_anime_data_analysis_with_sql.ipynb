```python
import pandas as pd
import sqlite3
```

```python
# SQLite veritabanı bağlantısını oluşturun
conn = sqlite3.connect('anime.db')
```

```python
# Veri setini okuyun ve SQLite veritabanına aktarın
data = pd.read_csv('/kaggle/input/imdb-rating-count-at-least-1000-anime/anime.csv')
data.to_sql('anime', conn, if_exists='replace', index=False)
```

```python
# SQL sorguları

#Tüm tabloya göz atalım
query = "Select * From anime"
tables = pd.read_sql(query, conn)
tables
```

```python
#Her türde kaç tane anime var?
query1 = "Select Genre, Count(*) as NumOfAnimes from anime group by Genre;"
tables1 = pd.read_sql(query1, conn)
tables1
```

```python
# Her anime türünün ortalama rating'i kaç?
query2 = "Select Genre, avg(Rating) as Avgrating from anime group by Genre order by Avgrating DESC;"
tables2 = pd.read_sql(query2, conn)
tables2
```

```python
# En yüksek ratinge sahip anime?
query3 = "SELECT Anime, Rating FROM anime WHERE Rating = (SELECT MAX(Rating) FROM anime)"
tables3 = pd.read_sql(query3, conn)
tables3
```

```python
# En yüksek ratinge sahip 10 anime?
query4 = "Select Anime, Rating From anime order by Rating DESC LIMIT 10;"
tables4 = pd.read_sql(query4, conn)
tables4
```

```python
# Yılda kaç anime yayınlandı?
query5 = "Select substring(Release_date, 1,5) AS Year, Count(*) As NumOfAnimes from anime group by substring(Release_date, 1,5) order by Year;"
tables5 = pd.read_sql(query5, conn)
tables5
```

```python
# Animelerin ortalama uzunluğu ne kadar?
query6 = "Select avg(cast(replace(length, ' min', '')AS FLOAT)) AS AvgLength from Anime"
tables6 = pd.read_sql(query6, conn)
tables6
```

```python
# Türe göre kategorize edilmiş, her yıl yayınlanan anime sayısı
query7 = "SELECT SUBSTRING(Release_date, 1, 5) AS Year, Genre, COUNT(*) AS NumOfAnimes FROM anime GROUP BY SUBSTRING(Release_date, 1, 5), Genre ORDER BY Year;"
tables7 = pd.read_sql(query7, conn)
tables7
```
